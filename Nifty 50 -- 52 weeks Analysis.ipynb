```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stock= pd.read_excel("NSE STOCKS RECENT .xlsx")
stock.head()
```

```python
stock.tail()
```

```python
stock.columns
```

```python
stocks=stock.drop(['SERIES', 'DATE OF LISTING','PAID UP VALUE', 'MARKET LOT', 'ISIN NUMBER'],axis=1)
stocks.head()
```

```python
stocks.dtypes
```

```python
stocks.isnull().sum()
```

```python
stocks.describe().T
```

```python
import seaborn as sns
sns.pairplot(stocks)
```

```python
print("Top 50 stockes based on 52 week high value ")
NF_HIGH=stocks.sort_values(by=["52 W - HIGH"],ascending=False).head(50).reset_index(drop=True)
NF_HIGH
```

```python
plt.figure(figsize=(30,15))
plots=sns.barplot(y="52 W - HIGH",x="NAME OF COMPANY",data=NF_HIGH.head(10))
for p in plots.patches:
    plots.annotate(format(p.get_height()),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center',size=23,
                   xytext=(0, 9),
                   textcoords='offset points')
plt.ylabel("52 - W HIGH",size=15)
plt.xlabel("Name of the company",size=20)
```
