# Most selling Product

```python
tc = df.groupby(['Product Code'])['Product Amount with GST'].sum().sort_values(ascending=False).nlargest(10).to_frame()
```

```python
tc = tc.reset_index()
tc
```

```python
sns.set(rc={"figure.figsize" :(15,10)})
ax = sns.barplot(data=tc,x='Product Code',y='Product Amount with GST')
ax.tick_params(axis='x',rotation=90)
```

# least selling Product

```python
sm = df.groupby(['Product Code'])['Product Amount with GST'].sum().sort_values(ascending=False).nsmallest(10).to_frame()
sm = sm.reset_index()
sm
```

```python
sns.set(rc={"figure.figsize" :(15,10)})
ax = sns.barplot(data=sm,x='Product Code',y='Product Amount with GST')
ax.tick_params(axis='x',rotation=90)
```
