```python
X = df.drop(['GDP (current US$) ','Year','Country Name'],axis=1)
y = df['GDP (current US$) ']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler ## standrard scalig 
scaler = StandardScaler() #initialise to a variable
scaler.fit(X_train) # we are finding the values of mean and sd from the td
X_train_scaled = scaler.transform(X_train) # fit (mean, sd) and then transform the training data
X_test_scaled = scaler.transform(X_test) # transform the test data 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train_scaled, y_train)
```

```python
regressor.intercept_ # c 
```

```python
regressor.coef_
```

```python
X_test_scaled
```

```python

y_pred = regressor.predict(X_test_scaled)

```

```python
df1= pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1
```

```python
from sklearn import metrics
train=regressor.predict(X_train_scaled)
r2_value=metrics.r2_score(y_train,train)
print("Training set R2_score  {}".format(r2_value))
```

```python
test=regressor.predict(X_test_scaled)
r2_value=metrics.r2_score(y_test,test)
print("Test set R2_score {}".format(r2_value))
```

```python
# or  we use this method to find the testing perfomance of the model above..
from sklearn import metrics
print('R2- SCORE:', metrics.r2_score(y_test,y_pred))
```

```python
MAE = metrics.mean_absolute_error(y_test,y_pred)
MSE = metrics.mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)

pd.DataFrame([MAE,MSE,RMSE],index=['MAE', 'MSE', 'RMSE'], columns=['Metrics'])
```

# cross validation the anove model

```python
from sklearn.model_selection import cross_val_score
lm_cv=LinearRegression()

scores = cross_val_score(estimator=lm_cv,X=X_train_scaled,y=y_train,cv=5) # 5 - fold cross validation
print("CV scores :",scores)
```

```python
from sklearn.model_selection import cross_val_predict
lm_cv_pt=LinearRegression()
cv_pred=cross_val_predict(estimator=lm_cv_pt, X=X_train_scaled,y=y_train,cv=5)
print("Accuracr score: ", metrics.r2_score(y_true=y_train, y_pred=cv_pred))
```
