```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("India_GDP_1960-2022.csv")
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.columns=df.iloc[0,:]
```

```python
df.sort_values(by=['Year'])
df.drop(0,inplace=True)
```

```python
df=df.sort_values(by=['Year'])
```

```python
df=df.astype({'Year':'int','GDP in (Billion) $': 'float', 'Per Capita in rupees': 'float','Growth %': 'float'})
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.boxplot(x=df['Growth %'],color='violet')
plt.show()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(35,20))
X=df["Year"]
Y=df["Growth %"]
Z=df["GDP in (Billion) $"]
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Y, 0.4, label = 'Growth %')
plt.bar(X_axis + 0.2, Z, 0.4, label = 'GDP in Billion')
  
plt.xticks(X_axis, X)

plt.xlabel("Year")
plt.ylabel("Growth")
plt.title("GDP in Billion")
plt.legend()
plt.show()
```

# Standard of Living

```python
plt.figure(figsize=(30,12))
sns.barplot(x="Year",y=" GDP per capita (current US$) ",data=df1)
```

This above graph shows the Standard of living ,GDP per capita is a popular measure of the standard of living, prosperity,
and overall well-being in a country. A high GDP per capita indicates a high standard of living, a low one indicates that a 
country is struggling to supply its inhabitants with everything they need. This Standared of Living increased fastly after LPG was implemented in the Indian Economy.

```python
plt.figure(figsize=(30,12))
sns.barplot(x="Year",y=" Total reserves (includes gold, current US$) ",data=df1)
```

When reserves increses - 
 increasing the (reserve requirement) ratios reduces the volume of deposits that can be supported by a given level of 
reserves and, in the absence of other actions, reduces the money stock and raises the cost of credit.

when reserves decreases - 
 it lowers the amount of cash that banks are required to hold in reserves, allowing them to make more loans to consumers and
businesses. This increases the nation's money supply and expands the economy.


Sudden increase and decrease of reserves - leads to Inflation.

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(35,20))
X=df1["Year"]
Y=df1["Imports of goods and services (% of GDP)"]
Z=df1["Exports of goods and services (% of GDP)"]
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Y, 0.4, label = 'Imports of goods and services')
plt.bar(X_axis + 0.2, Z, 0.4, label = 'Exports of goods and services')
  
plt.xticks(X_axis, X)

plt.xlabel("Year")
plt.ylabel("Imports of goods and services")
plt.title("Export of goods and services")
plt.legend()
plt.show()
```

# GDP growth % Annually

```python
plt.figure(figsize=(30,12))
sns.barplot(x="Year",y="GDP growth (annual %)",data=df1)
```


The ideal(Satisfactory) GDP growth rate is between 2% and 3%. The GDP growth rate measures how healthy the economy is. When the number is positive, the economy is growing. in india the GDP  grouth rate percentage is in negative for recent years because pandemic impacts (productivity issues).

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(35,20))
X=df1["Year"]
Y=df1["GDP growth (annual %)"]
Z=df1["Inflation, consumer prices (annual %)"]
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Y, 0.4, label = 'GDP growth')
plt.bar(X_axis + 0.2, Z, 0.4, label = 'Inflation on consumer prices')
  
plt.xticks(X_axis, X)

plt.xlabel("Year")
plt.ylabel("GDP growth")
plt.title("Inflation")
plt.legend()
plt.show()
```

The inflation and GDP  shows - government activity like war, sanctioned by other countries , closed economic policy affect the GDP of the country.

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=df1,x='Year',y='Inflation, consumer prices (annual %)',ax=ax,color='navy')
plt.show()
```

During later 90's the Inflation was increased heavyly because of the inpacts of sanctions fronm USA, followed by indo-china war and
